Importing the dependencies

```python
#importing the libraries
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import csv
```

Data collection and Data pre-processing

```python
#uploading the CSV file's contents to the added data frame
movies_data = pd.read_csv('/content/movies.csv')
```

```python
#printing a data frames's five rows
movies_data.head()
```

```python
#to determine the data frame's rows and columns counts
movies_data.shape
```

```python
#choosing the right features to suggest
selected_features = ['genres','keywords','original_language','tagline','cast','director',]
print(selected_features)
```

```python
#replacing the missing values with null string
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
#joining the all 5 rows
joined_features = movies_data['genres']+' ' +movies_data['keywords']+' '+movies_data['original_language']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
print(joined_features)
```

```python
#coverting the txt data in to feature vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(joined_features)
```

```python
#getting similarity score by using cosine similarity
similarity = cosine_similarity(feature_vectors)
print(similarity,"\n")
print(similarity.shape)
```

Getting the movie name from user

```python
#getting the name of the movie from the user
name_of_the_movie = input("Enter your favourite name of the movie:")
```

```python
#creating a list that contains all the name f the movie in tha given data set
all_titles = movies_data["title"].tolist()
print(all_titles)
```

```python
#finding the closest match for a name of the movie given by the user
close_match = difflib.get_close_matches(name_of_the_movie,all_titles)
print(close_match)
close_movie = close_match[0]
print(close_movie)
```

```python
#find the index of the movie with the tittle given by the user
index_of_the_movie = movies_data[movies_data.title == close_movie]["index"].values[0]
print(index_of_the_movie)
```

```python
#getting a similar score
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score,"\n")
len(similarity_score)
```

```python
#based on their similarity score for sorting the movies
similarity_score_sorted = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print(similarity_score_sorted)
```

```python
#based on their idex to print the name of similar movies
print("movies suggested for you:")
i = 1
for movie in similarity_score_sorted:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]["title"].values[0]
  if i<31:
    print(i,'.',title_from_index)
    i = i+1
```

Movie Recommendation

```python
name_of_the_movie = input("Enter your favourite name of the movie:")
all_titles = movies_data["title"].tolist()
close_match = difflib.get_close_matches(name_of_the_movie,all_titles)
close_movie = close_match[0]
index_of_the_movie = movies_data[movies_data.title == close_movie]["index"].values[0]
similarity_score = list(enumerate(similarity[index_of_the_movie]))
similarity_score_sorted = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print("movies suggested for you:")
i = 1
for movie in similarity_score_sorted:
  index = movie[0]
  title_from_index = movies_data[movies_data.index == index]["title"].values[0]
  if i<31:
    print(i,'.',title_from_index)
    i = i+1
```
